```python
import pandas as pd
import numpy as np
```

```python
#1. Load the csv file
df = pd.read_csv("IITK_Courses_Report.csv")
```

```python
df.head()
```

```python
#2. Check for null values in all columns
df.isnull().sum()
```

```python
#3. For rows where either of the non-numerical columns contain null values, delete the entire row
df_new = df.dropna(subset = ["Student Name", "Course"]).copy()
df_new
```

```python
df_new.isnull().sum()
```

```python
#4. For numerical columns, replace each null value with the geometric mean of the values in the rows immediately above and below the null value.
for i in range(1, len(df_new)-1):
    if pd.isnull(df_new["Marks (%)"].iloc[i]):
        prev_val = df_new["Marks (%)"].iloc[i-1]
        next_val = df_new["Marks (%)"].iloc[i+1]

        mean = np.sqrt(prev_val * next_val)
        df_new.at[df_new.index[i], "Marks (%)"] = mean
```

```python
for i in range(1, len(df_new)-1):
    if pd.isnull(df_new["Attendance (%)"].iloc[i]):
        prev_val = df_new["Attendance (%)"].iloc[i-1]
        next_val = df_new["Attendance (%)"].iloc[i+1]

        mean = np.sqrt(prev_val * next_val)
        df_new.at[df_new.index[i], "Attendance (%)"] = mean
```

```python
df_new.to_csv('cleaned_IITK_Courses_Report.csv', index=False)
```
